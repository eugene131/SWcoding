### list 문제 1번
- 다음과 같은 list1, list2가 있을 경우,
    - list1,list2 생성
- 이중 for 루프를 사용하여
    - 첫번째 for문의 i는 list1의 리스트를 하나씩 받아오고, 두번째 for문의 j는 list2의 리스트를 하나씩 받아옴. 
    - 이 때, i가 하나 돌 동안 j는 모두 한바퀴 다 돌기 때문에 i가 끝까지 돌고 나면 모든 경우의 수가 나옴 
- list1과 list2의 각 원소를 곱한 후
- 원소의 곱셈을 결과와 함께 출력하시오.
```
list1 = [3,5,7]
list2 = [2,3,4,5,6]
```

````python
list1=[3,5,7]
list2=[2,3,4,5,6]
for i in list1:
    for j in list2:
        print(i,'*',j,'=',i*j)
````

### list문제 2번
- a = [2, 3, 4, 5, 6]이 있을 경우, 
- 이 리스트의 순서를 바꾸는 기능을 
- for -in 문과 pop() 메소드를 사용하여 구현하시오
    - for -in으로 len(a)의 수만큼 돌면서
- (reverse() 메소드를 사용하지 않고 리스트의 원소를 하나하나 순회하면서
- pop() 메소드를 호출하시오)
    - pop()메소드를 호출한다.
    - pop()메소드는 맨 뒤에 있는 수부터 차례로 뽑아 내고 삭제하기 때문에 리스트의 길이만큼 pop()을 호출하면 정확히 역순으로 데이터가 추출됨

````python
a=[2,3,4,5,6]
rev_a=[]
print('a =',a)
for i in range(len(a)):
    rev_a.append(a.pop())
print('rev_a =',rev_a)
````

### list 문제 8번
- s_list = [‘abc, ‘bcd’, abc’, ‘abba’, ‘cddc’, ‘opq’, ‘opq’] 와 같은 문자열을 가진 리스트가 존재한다. 
- 이 리스트에서 중복해서 나타나는 원소들을 제거한 리스트 new_list = [‘abc’, ‘bcd’, ‘abba’, ‘cddc’, ‘opq’]를 생성하여라. 
    - s_list에서 요소를 하나씩 추출하는 for문을 생성한다
    - new_list에 하나씩 추가하며 not in 커맨드를 사용해 내부에 그 str이 없는 경우에만 append한다

````python
s_list =  ['abc', 'bcd' , 'abc', 'abba','cddc', 'opq', 'opq']
new_list=[]
for i in s_list:
    if i not in new_list:
        new_list.append(i)
new_list
````

### 딕셔너리, 튜플, 셋 6번 문제
- 주어진 튜플은 10일 동안 일일 매장 매출을 기록한 것이다.
- 매출이 전날보다 감소한 날이 며칠인지 출력하는 코드를 작성하라.
    - 뒷 요소에서 앞 요소를 뺀 값이 음수이면 카운트를 하나씩 올려주면 되는 문제
```
tu=(100,121,120,130,140,120,122,123,190,125)
```

````python
tu=(100,121,120,130,140,120,122,123,190,125)
count=0
for i in range(1,len(tu)):
    if 0>tu[i]-tu[i-1]:
        count=count+1
print('지난 '+str(len(tu))+'일 동안 전일대비 매출이 감소한 날은 '+ str(count)+'일입니다.')
````

### 딕셔너리, 튜플, 셋 3번 문제
- 어떤 문자열을 뒤집었을 때에 원래의 문자열과 같은 것을 회문(palindrome)이라고 한다. 
- 예를 들어 ‘mom’은 문자열을 뒤집어도 ‘mom’이 되므로 회문이다. 
- 사용자로부터 임의의 문자열이나 문장을 입력받아서
- 이 문자열이 회문인지 아닌지 검사하는 코드를 작성하라. 
- 입력으로는 다음과 같은 영문자를 입력으로 받으며, 
- 대문자와 소문자, 공백, 마침표 등은 무시하도록 구현하라.
    - 이 부분이 조금 어려웠음
    1. 알파벳을 소문자 or 대문자로 통일한다.
    2. 공백과 여러 특수 문자를 지운다
        - 특수문자 삭제를 위해 replace문을 이용했다
        - for문으로 특수문자를 모아둔 리스트에 접근해서
        - 요소 하나하나를 모두 집어 넣어 replace를 통해 모두 지웠다
    3. 문자열을 인덱싱을 이용해 거꾸로 만들어 주고
    4. 둘을 비교해서 같으면 회문, 다르면 회문이 아니다.

````python
stri=input('문자열을 입력하시오 : ')
punctuation ='!"$%&\' ()*+,-./:;<=>?[\]^_{|}~'
stri=stri.lower()#1번 내용
for i in range(len(punctuation)):#2번 내용
    stri=stri.replace(punctuation[i],'')
new_stri=stri[-1::-1]#3번 내용
if stri==new_stri:#4번 내용
    print('회문입니다.')
else:
    print('회문이 아닙니다.')
````
