```python
def divide(x,y):
    try:
        result = x/y
    except ZeroDivisionError:
        print('0으로 나누는 오류')
    else:
        print('결과:',result)
    finally:
        print('수행완료')
print('divide(100,2) 함수호출:')
divide(100,2)
print('divide(100,0) 함수호출:')
divide(100,0)
```

```python
def get_ans(ans):
    if ans in ['예','아니요']:
        print('정상적인 입력입니다.')
    else:
        raise ValueError("입력을 확인하세요.")
while True:
    try:
        ans = input("예/아니요 중 하나를 입력하세요:")
        get_ans(ans)
        break
    except Exception as e:
        print('error',e)
```

```python
def print_file_word(path):
    fi = open(path,'r')
    while True:
        s=fi.readline()
        if s=='':
            break
        print(s,end='')

f= open('foo.txt','w')
f.write('AAA\nBBB\nCCC\n')
f.close()
print_file_word('foo.txt')
print('='*10+'after append'+'='*10)
f = open('foo.txt','a+')
f.write('use append mode\n')
f.close()
print_file_word('foo.txt')
```

```python
import sys
fname = input('입력할 파일 이름: ')
try:
    f= open(fname,'r',encoding='UTF8')
except IOError:
    print('Could not read file:',fname)
    sys.exit()
except:
    print('Unexpected error:',sys.exc_info()[0])
    sys.exit()
n=1
I=f.readline()
while I:
    print('{:3d}:{}'.format(n,I),end = '')
    n+=1
    I=f.readline()
f.close()
```

```python
def print_save_file_upper(path):
    fi = open(path,'r')
    f = open(path.replace('.txt','')+'_upper.txt','w')
    while True:
        s=fi.readline()
        if s=='':
            break
        f.write(str.upper(s))
        print(str.upper(s),end='')

fi_name=input('입력할 파일의 이름 :')
print_save_file_upper(fi_name)

```

```python
def save_list_file_world(path):
    fi = open(path,'r')
    fi_list=[]
    while True:
        s=fi.readline()
        if s=='':
            fi.close()
            break
        fi_list.append(s)
    return fi_list

fi_name=input('입력할 파일의 이름 :')
fi_list=save_list_file_world(fi_name)
for i in fi_list[::-1]:
    print(i,end='')
```
