3. 위의 randint30.txt라는 파일을 읽어서 1에서 10까지 정수의 출현횟수를 다음과 같이 출력하여라.
image.png
    - 정수가 몇개 있는지 각각 세어 줄 필요가 있다. 
        - 공백으로 나눠진 수를 읽어 리스트에 넣어준다.
        - 리스트를 읽어주며 1이라는 수가 나오면, count리스트를 만들어, 위치에 해당하는 count 수를 1씩 올려주는 방법을 이용하여 카운트 한다. 

```python
import random as rd

def count_num_1_10(n_list):
    count=[]
    for i in range(10):
        count.append(0)
    for i in n_list:
        count[i-1]+=1
    return count
num_list=[]
for i in range(30):
    num_list.append(rd.randrange(1,11))
f=open('randint30.txt','w')
for i in num_list:
    f.write(str(i)+' ')
f.close()
f=open('randint30.txt','r')
r_f_num=list(map(int,f.read().split()))
count_nums=count_num_1_10(r_f_num)
for i in range(len(count_nums)):
    print('{} : {}개'.format(i+1,count_nums[i]))
```
