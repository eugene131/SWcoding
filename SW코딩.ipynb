### 반지름을 입력 받아 원의 넓이를 구한는 문제
- 원의 넓이를 구하는 공식은(반지름 * 반지름 * 파이)로 구할 수 있다.
- 다음의 r은 반지름 파이는 3.14로 나타냈기 때문에 결국 구하는 식은 r의 제곱 곱하기 3.14 이고 파이썬으로 표현하면 위의 식과 같다.
