## 1번 문제 
- 로미오와 줄리엣 두 사람이 주사위를 던져서 높은 숫자가 나오면 이기는 게임을 만들어보자. 
- 로미오와 줄리엣의 주사위는 모두 다음과 같이 random 모듈의 randrange( )를 통해서 생성한 난수를 바탕으로 한다. 
    - 즉, random 패키지를 import해줘야 함
    - randrange(a,b)함수는 입력된 범위(a<=n<b)만큼 리턴함-> 주사위의 눈은 1~6까지 있기 때문에 1,2,3,4,5,6이 리턴되도록 범위를 1,7로 지정
- 출력 결과는 다음과 같이 로미오가 이기거나, 줄리엣이 이기거나, 비기는 결과가 나와야 한다.
    - if문을 통해 로미오가 이기는 경우, 지는 겨우, 비기는 경우를 나누어 출력한다.

```python
import random as rd
rom=rd.randrange(1,7)
print("로미오의 주사위는 {}입니다".format(rom))
jul=rd.randrange(1,7)
print("줄리엣의 주사위는 {}입니다".format(jul))
if rom>jul:
    print("로미오가 이겼습니다.")
elif jul>rom:
    print("줄리엣이 이겼습니다.")
else:
    print("비겼습니다.")
```

## 2번 문제
- 랜덤 숫자 맞추기 게임 프로그램을 작성하라. 
- 먼저 랜덤하게 1~20까지의 숫자 x를 하나 생성시키고
    - random 모듈을 사용해야 함
    - randrange메서드를 사용해서 범위를 (1,21)로 지정해서 1부터 20까지의 난수가 발생하도록 함
- 사용자는 숫자를 하나씩 입력하면서 생성된 숫자 x와 비교해 큰지 작은지를 보고 숫자를 맞춰가는 게임이다.
    - 사용자로부터 input을 받아 int형으로 변환시켜 x와 비교해준다
- 컴퓨터는 사용자가 입력한 숫자가 정답보다 큰지 작은지를 알려주어야 한다.
    - 크거나 작은 상황을 if문으로 비교하여 어떤 상황인지 출력해 준다.
- 그리고 사용자가 시도를 할 때 마다 시도한 횟수가 n에 저장되도록 한다.
    - input이 들어올 때마다 n을 1씩 증가시켜 준다
- 만일 입력한 숫자와 x가 일치하면 “정답입니다!”라는 메시지를 출력한다.
    - input과 x를 비교하다가 같은 것이 나오면 "정답입니다!"를 출력
- 이때 3번 이하로 입력하여 숫자를 맞추면 “n번 만에 맞춘 당신은 천재!”, 3번 이상 6번 이하로 입력하여 숫자를 맞추면 “n번 만에 맞추셨네요. 잘 했어요^^”, 7번 이상으로 숫자를 입력하여 맞추면 “n번 만에 맞추다니 쩝쩝…”이라고 출력하라.
    - 1씩 증가시키던 n을 이용하여 if문으로 위 조건에 맞는 숫자를 비교해 출력해 준다. 

```python
import random as rd
x=rd.randrange(1,21)
n=0
while True:
    num=int(input('1~20까지의 숫자를 입력하세요:'))
    n+=1
    if num==x:
        print("정답입니다.")
        break
    elif num<x:
        print("{}보다 큽니다.".format(num))
    else:
        print("{}보다 작습니다.".format(num))
if n<=3:
    print('{}번만에 맞춘 당신은 천재!'.format(n))
```

## 6번 문제
- 춘향이와 몽룡이는 2019년 2월 24일 발렌타인데이에 연애를 시작했다.
    - 연애를 시작한 날짜에 위 날짜를 저장해 준다.
- 오늘은 두사람의 연애 시작일로부터 며칠이 경과하였는가?
    - 오늘 날짜에서 연애 시작 날짜를 빼준다.
- Datetime 모듈을 사용하고 오늘날짜 date.today()를 통해 받아와서
    - 오늘 날짜를 받아올 때, datetime을 import해주고, datetime 패키지의 date클래스에서 today매서드를 사용해 준다.
- 오늘날짜, 연애를 시작한 날짜와 경과한 날짜, 기념일을 다음과 같이 출력하여라.
    - 각 날짜를 출력해 주는데, 기념일 같은 경우 timedelta메서드를 사용하여 지낙 날짜를 구하고, 출력해 준다.  

```python
import datetime as dt
today=dt.date.today()
start_day=dt.date.today().replace(year=2019,month=2,day=24)
days_gap=today-start_day
avg_100=today+dt.timedelta(days=100)
avg_200=today+dt.timedelta(days=200)
avg_500=today+dt.timedelta(days=500)
avg_1000=today+dt.timedelta(days=1000)
print("오늘은 {}년 {}월 {}일입니다".format(today.year,today.month,today.day))
print("춘향이와 몽룡이의 연애 시작일 : {}년 {}월 {}일".format(start_day.year,start_day.month,start_day.day))
print("연애 시작일로부터 경과한 날짜: {}일".format(days_gap.days))
print("100일 기념일 : {}년 {}월 {}일".format(avg_100.year,avg_100.month,avg_100.day))
print("200일 기념일 : {}년 {}월 {}일".format(avg_200.year,avg_200.month,avg_200.day))
print("500일 기념일 : {}년 {}월 {}일".format(avg_500.year,avg_500.month,avg_500.day))
print("1000일 기념일 : {}년 {}월 {}일".format(avg_1000.year,avg_1000.month,avg_1000.day))

```

## 7번 문제
- 1부터 1,000,000까지 정수의 합을 구하여 반환하는 함수 sum1to1000000()을 작성하고 
    - 1부터 1,000,000까지 구하는 것은 0부터 10^6까지 더하는 합과 같다.
    - 이것을 리턴해 주면 된다. 
- 함수를 100번 호출하여 함수 수행 시간을 다음과 같이 초단위로 구하여 출력하여라. 
    - 함수를 호출하기 전, time모듈을 사용하여 호출 전의 시간(a)을 측정해준다.
    - 함수 호출이 모두 끝난 후, time모듈을 사용하여 호출 후의 시간(b)을 측정해 준다.
    - b-a를 해 준다.
- (소수점 4자리까지 표현할 것.)
    - 출력 형식을 맞추기 위해 format을 사용하여 아해 코드와 같이 나타낸다. 

```python
import time
def sum1to1000000():
    return sum(range(1000001))

start_time=time.time()
for _ in range(100):
    sum1to1000000()
end_time=time.time()
print('1에서 1,000,000까지의 합을 100번 반복해서 구하는 시간 : {:.4f}초'.format(end_time-start_time))
```
