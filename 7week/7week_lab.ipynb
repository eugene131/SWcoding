## 구두과제


### 1. 주사위 게임
- 사람 a와 사람 b가 주사위 게임을 한다.
- randInt를 사용하여 a가 이겼는지 b가 이겼는지 결과를 출력하라 
    - 단, 무승부일 경우 한쪽이 승리할 때까지 다시 승부하도록 한다.

```python

while True:
    dice_a=rd.randint(1,6)
    dice_b=rd.randint(1,6)
    if dice_a==dice_b:
        continue
    if dice_a>dice_b:
        print("a win \ndice a : {}, dice b : {} ".format(dice_a,dice_b))
        break
    else:
        print("b win \ndice a : {}, dice b : {} ".format(dice_a,dice_b))
        break
```

### 2. 로또 넘버 만들기
- 1~45의 수를 랜덤하게 출력하여 list에 6개를 담아 준다.
- 이 리스트를 출력한다.

```python
def roto_num_zener():
    roto_list=[]
    for i in range(1,46):
        roto_list.append(i)
    rd.shuffle(roto_list)
    return roto_list[:6]
print(roto_num_zener())
```
