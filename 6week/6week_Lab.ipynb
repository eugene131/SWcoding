### LAB 4-6

```python
def print_root(a,b,c):
    r1=(-b+(b**2-4*a*c)**0.5)/(2*a)
    r2=(-b-(b**2-4*a*c)**0.5)/(2*a)
    print("해는",r1,"또는",r2)
```

```python
#1번-(1)
print_root(1,4,-21)
```

```python
#1번-(2)
print_root(1,-6,8)
```

### LAB 4-7

```python
def circle_area_circum(radius):
    area = radius ** 2 * 3.14
    circum = radius*2*3.14
    print('반지름 %d인 원의 면적은 %.1f 원의 둘레는 %.1f'%(radius,area,circum))
    return area,circum
area, circum = circle_area_circum(10)    
```

### LAB 4-8

```python
def multiplies(n,m):
    a=(n,)
    for i in range(1,m):
        b=(n*(i+1),)
        a=a+b
    return a
r1,r2,r3,r4 = multiplies(3,4)
print(r1,r2,r3,r4)
r1,r2,r3,r4,r5 = multiplies(2,5)
print(r1,r2,r3,r4,r5)
```

### LAB 4-10

```python
#1번
def sum_nums(*numbers):
    result = 0
    print('%d개의 인자'%len(numbers),numbers)
    for n in numbers:
        result+=n
    avr= result/len(numbers)
    return result,avr
result,avr = sum_nums(10,20,30)
print("합계 :",result,", 평균 :",avr)
result,avr = sum_nums(10,20,30,40,50)
print("합계 :",result,", 평균 :",avr)
```
