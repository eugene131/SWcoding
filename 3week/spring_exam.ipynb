#### 1번문제
섭씨, 화씨 변환
- 섭씨온도(Celsius)를 화씨온도(Fahrenheit)로 변환하는 식은 아래와 같다. Fahrenheit = (9/5) * Celsius + 32 이 식을 바탕으로 섭씨온도를 0도에서 50도까지 5도 단위로 증가시키면서 이에 해당하는 화씨온도를 계산하여, 섭씨온도와 화씨온도의 값을 보여주는 프로그램을 작성하시오
    1. (1)단순히 식 대입 문제 + (2)0도에서 50도까지 5도씩 증가시키는 문제
        1. 식을 그대로 적어주기만 하면 되는 문제
        2. for문 과 range를 통해 0도 부터 50도까지 나올 수 있게 구현 그리고 그 안에 (1)을 넣어주면 됨 

```python

for C in range(0,51,5):
    F=(9/5) * C +32
    print('섭씨 : %5dC'%C,'|','화씨 : %5dF:'%F)
```

#### 3번 문제
- 10 미만의 자연수에서 3과 5의 배수를 구하면 3, 5, 6, 9이다. 이들의 총합은 23이다. 1000 미만의 자연수에서 3의 배수와 5의 배수의 총합을 구하시오.
    - 푸는 방법은 2가지
    1. 애초에 3과 5의 공배수가 하나만 들어가게 하거나
        - (a) 3으로 나눈 나머지가 0이면 합계에 더해 준다
        - (b) 5로 나눈 나머지가 0이면 합계에 더해 주는데,
        - (c) elif를 사용하여 (a)의 경우 더해진 것를 제외한 (b)의 경우만 더해준다. 
    2. 겹치게 구한 후 15의 배수를 빼주는 방법
        - (a) 3으로 나눈 나머지가 0이면 합계에 더해 준다
        - (b) 5로 나눈 나머지가 0이면 합계에 더해 준다
        - (c) (a)+(b)의 값- 15로 나눈 나머지가 0인 수의 합계



```python
sum=0
for i in range(3,1001):
    if i%3==0:
        sum=sum+i
    elif i%5==0:
        sum=sum+i
print(sum)
```
