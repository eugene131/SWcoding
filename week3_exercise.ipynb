### 5번문제
1. 소수란 양의 자연수 중에서 1과 자기 자신이외의 약수를 가지지 않는 수를 말한다. 
    - 구해야 할 수의 정의 즉 자기 자신까지의 수를 전부 나누었을때 나머지가 0인 것이 없어야 함(%연산 사용해야 하는 것을 알 수 있음)
2. 사용자로부터 양의 정수 n을 입력받은 후
    - (생략)
3. 이 숫자가 소수인지 아닌지를 판별하는 프로그램을 작성하시오. 
    - 위의 1.부분을 참고하면 2부터 자기자신-1이 될 때까지, 나머지가 0이 있는지 찾아야 함
4. 이때 수행 속도를 개선하기 위하여 2로 나누어 본 후 나누어지지 않을 경우 3, 5, 7, … 과 같은 홀수로 나눗셈을 시도하도록 코딩하여라.
    - 2로 나눈 나머지를 구해 보면 짝수인지 아닌지 알 수 있는데, 짝수이면 소수가 아니므로 더 구할 필요가 없다.즉, 짝수가 아닐경우 3부터 시작해서 2씩 증가시키며 나머지를 구해 본다.
-	힌트: n을 2부터 n-1까지의 정수로 모두 나누어 본다. 이 때 나누어 떨어지는 원소가 하나라도 있으면 소수가 아니다.


````python
n=int(input('숫자를 입력하세요 : '))
if n==1:
    print(n,'은 소수가 아닙니다')
elif n==2:
    print(n,'은 소수입니다')
else:    
    if (n%2) != 0:#짝수인지 확인, 짝수가 아니라면,
        for i in range(3,n,2):#3부터 n-1까지 2씩 증가시키며 홀수를 나눠봄
            if (n%i) == 0:#나누어 떨어지면, n은 소수가 아님
                print(n,'은 소수가 아닙니다')
                break
        else:#for문이 돌면서 나누어 떨어지는게 없으면(break가 안걸리면) 소수
            print(n,'은 소수입니다')
    else:#짝수이면 소수가 아님
        print(n,'은 소수가 아닙니다')
    
````

### 6번 문제	
- 암스트롱 수란 세 자리의 정수 중에서 각 자리의 수를 세제곱한 수의 합과 자신이 같은 수를 말한다. 구체적인 예를 들면 153은 1 + 125 + 27 의 합으로 구성될 수 있는데 이러한 수가 암스트롱 수이다. 세 자리 정수들 중에서 모든 암스트롱 수를 구하여 다음과 같이 출력하여라.



````python
print('세 자리의 암스트롱 수 : ',end='')
for i in range(100,1000):
    num1=i//100
    num2=(i%100)//10
    num3=i%10
    if (num1**3)+(num2**3)+(num3**3)==i:
        print(i,end=' ')
````

### 9번 문제
- 이중 for 문을 사용하여 숫자를 입력 받아 다음과 같은 삼각형을 출력하는 프로그램을 작성하시오. 이때 아래 결과와 같이 숫자 10이 입력되면 높이가 10이고 제일 마지막 줄의 별이 10개가 삼각형 형태로 나타나도록 하시오.
- 힌트: 첫번째줄은 공백 9개, 별1개 / 마지막줄은 공백이 없다. 


````python
n=int(input('숫자를 입력하세요 : '))
for i in range(1,n+1):
    for j in range(n-i):
        print(' ',end='')
    for j in range(i):
        print('*',end='')
    print()
````

### 10번 문제
1. 사용자로부터 숫자 1보다 크고 10보다 작은 값 n을 입력으로 받아서
    - (생략)
2. 다음과 같이 뱀의 몸통처럼 증가하는 이차원 배열을 출력하여라.
    - 아래의 참조를 통해 ㄹ과 같은 패턴을 그림을 알 수 있음
    - 즉 다르게 말하면, (a)첫 라인은 왼쪽에서 오른쪽으로 증가,(0부터 시작하기 때문에, 짝수 라인)
    - (b)그 다음 라인은 오른쪽에서 왼쪽으로 증가,(홀수 라인)
        - 다르게 말하면 왼쪽에서 오른쪽으로 감소
        - 그리고 시작 값은 (a)의 맨 오른쪽 값에서 n을 더한 값과 같음
            - *한 줄이 증가할 때 최대 수는 맨 오른쪽에서 n을 더한 값과 같기 때문에*
    - (a),(b)를 이용하여 차례로 리스트를 출력
- (이 배열은 마치 뱀의 몸통처럼 ‘ㄹ’과 같은 패턴을 그리며 수가 1씩 증가하는 형태의 배열이다.) 

````python
n=int(input('n을 입력하시오 : '))
run_num=1
for i in range(n):
    
    if i%2==0:
        for j in range(n):
            print('%4d'%(run_num),end='')
            run_num=run_num+1
    else:
        run_num=run_num+n
        for j in range(n):
            print('%4d'%(run_num-(j+1)),end='')
    
    print()
````

#### 3번 문제
- 10 미만의 자연수에서 3과 5의 배수를 구하면 3, 5, 6, 9이다. 이들의 총합은 23이다. 1000 미만의 자연수에서 3의 배수와 5의 배수의 총합을 구하시오.
    - 푸는 방법은 2가지
    1. 애초에 3과 5의 공배수가 하나만 들어가게 하거나
        - (a) 3으로 나눈 나머지가 0이면 합계에 더해 준다
        - (b) 5로 나눈 나머지가 0이면 합계에 더해 주는데,
        - (c) elif를 사용하여 (a)의 경우 더해진 것를 제외한 (b)의 경우만 더해준다. 
    2. 겹치게 구한 후 15의 배수를 빼주는 방법
        - (a) 3으로 나눈 나머지가 0이면 합계에 더해 준다
        - (b) 5로 나눈 나머지가 0이면 합계에 더해 준다
