```python
#1번
a = np.arange(0,24).reshape(4,3,2)
print("index \t\t element")
for x1 in range(0,4):
    for x2 in range(0,3):
        for x3 in range(0,2):
            print("({0},{1},{2})  \t  {3}".format(x1,x2,x3,a[x1,x2,x3]))
```

```python
#2번
print(np.concatenate((a[0,0],a[1,0]),axis=0))
print(np.concatenate((a[0,0],a[2,1]),axis=0))
print(np.concatenate((np.concatenate((a[0,0],a[1,0]),axis=0),np.concatenate((a[0,1],a[1,1]),axis=0),np.concatenate((a[0,2],a[1,2]),axis=0)),axis=0).reshape(3,4))
```
