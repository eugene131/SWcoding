```python
class OddCounter:
    def __init__(self, n=1):
        self.n=n
    def __iter__(self):
        return self
    def __next__(self):
        if self.n<20:
            t=self.n
            self.n+=2
            return t+self.n
        raise StopIteration
```

```python
my_cout = OddCounter()
print(next(my_cout))
print(my_cout.__next__())
```

```python
#1번 
class EvenCounter:
    def __init__(self,n=0):
        self.n=n
    def __iter__(self):
        return self
    def __next__(self):
        if self.n<=20:
            t=self.n
            self.n+=2
            return t
        raise StopIteration

my_even=EvenCounter()

for i in range(6):
    print(next(my_even), end=' ')
print()
#2번
my_even.__init__(0)
for i in my_even:
    print(i,end =' ')
```
