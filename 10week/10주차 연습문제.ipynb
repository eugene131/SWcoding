```python
class Counter:
    def __init__(self,number=0):
        if number>=100 | number<=-1:
            self.__number=0
        self.__number=number
    def reset(self):
        self.__number=0
    def inc(self):
        self.__number+=1
        if self.__number>=100:
            self.reset()
    def dec(self):
        self.__number-=1
        if self.__number<=-1:
            self.reset()
    def __str__(self):#100이상 이면0 으로
        return "C({})".format(self.__number)
    def __add__(self,other):
        if self.__number+other.__number>=100:
            result = 0
        else:
            result = self.__number+other.__number
        return 'C({})'.format(result)
    def __sub__(self,other):
        if self.__number-other.__number<=-1:
            result = 0
        else:
            result = self.__number-other.__number
        return 'C({})'.format(result)    
c1=Counter(10)
c1.inc()
print('c1 =',c1)

print()
c2=Counter()
c2.inc()
c2.inc()
c2.dec()
print('c2 =',c2)
c2.reset()
print('c2 =',c2)
c1 = Counter(10)
c2 = Counter(20)
c3 = c1+c2
c4 = c1-c2
print('c3 =',c3)
print('c4 =',c4)
```

```python
class BankAccount:
    def __init__(self, __name, __account_num):
        self.__name = __name
        self.__account_num=__account_num
        self.__balance=0
    def __str__(self):
        return "{}님의 계좌 {}의 잔고는 {}원입니다.".format(self.__name,self.__account_num,self.__balance)
    def get_name(self):
        return self.__name
    def get_account_num(self):
        return self.__account_num
    def get_balance(self):
        return self.__balance
    def deposit(self, money):
        self.__balance = money + self.__balance
        print("{}원이 입금되었습니다. 잔고는 {}원입니다.".format(money,self.__balance))
    def withdraw(self, money):
        if self.__balance-money<0:
            print("계좌 잔고는 {}원으로 인출 요구 금액 {}원보다 작습니다.".format(self.__balance,money))
        else:
            self.__balance = self.__balance - money
account1 = BankAccount('홍길동', '1234-0001')
print(account1)
account1.deposit(2000)
print(account1)
account1.withdraw(500)
print(account1)
account1.withdraw(5000)
```

```python
class Student:
    def __init__(self,__name,__student_id):
        self.__name=__name
        self.__student_id=__student_id
        self.__korean_quiz=0
        self.__math_quiz=0
        self.__science_quiz=0
        self.__total_score=0
    def __str__(self):
        return ("이름:{}, 학번:{} 국어성적:{}, 수학성적:{}, 과학성적:{}, 합계:{}, 평균:{}".format(self.__name,self.__student_id,self.__korean_quiz,self.__math_quiz, self.__science_quiz, self.__total_score, self.__total_score/3))
    def get_name(self):
        return self.__name
    def get_student_id(self):
        return self.__student_id
    def get_korean_quiz(self):
        return self.__korean_quiz
    def get_math_quiz(self):
        return self.__math_quiz
    def get_science_quiz(self):
        return self.__science_quiz
    def set_korean_quiz(self,score):
        self.__korean_quiz=score
        self.__total_score=self.__total_score + score
    def set_math_quiz(self, score):
        self.__math_quiz=score
        self.__total_score=self.__total_score + score
    def set_science_quiz(self, score):
        self.__science_quiz=score
        self.__total_score=self.__total_score + score
    def get_total_score(self):
        self.__total_score=self.__science_quiz+self.__math_quiz+self.__korean_quiz
        return self.__total_score
    def get_avg_score(self):
        return (self.__science_quiz+self.__math_quiz+self.__korean_quiz)/3
name = input('학생의 이름을 입력하세요 : ')
student_id = input('학생의 학번을 입력하세요 : ')

student = Student(name,student_id)

korean_quiz = int(input('학생의 국어 성적을 입력하세요 : '))
math_quiz = int(input('학생의 수학 성적을 입력하세요 : '))
science_quiz = int(input('학생의 과학 성적을 입력하세요 : '))
student.set_korean_quiz(korean_quiz)
student.set_math_quiz(math_quiz)
student.set_science_quiz(science_quiz)
print(student)
```

```python
class TV:
    MIN_VOLUME=0
    MAX_VOLUME=20
    MIN_CHANNEL=0
    MAX_CHANNEL=200

    def __init__(self):
        self.__volume=5
        self.__channel=0
        self.__is_on=False
    def __str__(self):
        if(self.__is_on):
            return "불륨 = {}, 채널 = {}".format(self.__volume,self.__channel)
        else:
            return "TV가 꺼짐 상태입니다"
    def toggle_power(self):
        if self.__is_on==False:
            self.__is_on=True
        else:
            self.__is_on=False
    def get_channel(self):
        return self.__channel
    def set_channel(self,choice):
        if 0>=choice | choice>=201: 
            print("채널오류")
        else:
            self.__channel = choice
    def get_volume(self):
        return self.__volume
    def set_volume(self, choice):
        if 0<choice<20: 
            print("볼륨오류")
        else:
            self.__volume = choice
    def volume_up(self):
        global MAX_VOLUME
        if self.__volume > MAX_VOLUME: 
            self.__volume=MAX_VOLUME
        else:
            self.__volume+=1
    
    def volume_down(self):
        global MIN_VOLUME
        if self.__volume<MIN_VOLUME: 
            self.__volume=MIN_VOLUME
        else:
            self.__volume-=1
    def channel_up(self):
        global MAX_CHANNEL
        if self.__channel>=MAX_CHANNEL: 
            self.__channel=0
        else:
            self.__channel+=1
    def channel_down(self):
        global MIN_CHANNEL
        if self.__channel<MIN_CHANNEL: 
            self.__channel=MIN_CHANNEL
        else:
            self.__channel-=1

my_tv=TV()
print(my_tv)
my_tv.toggle_power()
print(my_tv)  
my_tv.set_channel(200)
print(my_tv)
my_tv.volume_up()
my_tv.channel_up()
print(my_tv)
```
