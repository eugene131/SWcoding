## 3. 실행시간이 오래 걸리기 때문에 쓰레드로 실행

```python
subject = ['manage','database','mobile','SWcoding','dataS','comp_struct']

thr=[]
thr.append(Thread(target=manage,args=(driver[0],)))
thr.append(Thread(target=database,args=(driver[1],)))
thr.append(Thread(target=mobile,args=(driver[2],)))
thr.append(Thread(target=SWcoding,args=(driver[3],)))
thr.append(Thread(target=dataS,args=(driver[4],)))
thr.append(Thread(target=comp_struct,args=(driver[5],)))

for i in range(6):
    thr[i].start()
for i in range(6):
    thr[i].join()
```
