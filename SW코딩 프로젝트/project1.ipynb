## 컴퓨팅사고와 SW코딩 수업으로 접근
- 강의자료실에 업로드
- 드라이버 3 사용

```python
def SWcoding(driver):
    driver.get('https://canvas.knu.ac.kr/')
    driver.implicitly_wait(10)
    
    #SW코딩 강의로 접근
    driver.find_element(By.XPATH,'//*[@id="DashboardCard_Container"]/div/div[7]/div/a').click()
    print("강의로 접근중...")
    driver.implicitly_wait(10)
    print("강의로 접근완료")
    #강의자료실 클릭릭
    driver.find_element(By.XPATH,'//*[@id="section-tabs"]/li[6]/a').click()
    print("강의 자료실로 접근중...")
    driver.implicitly_wait(10)
    print("강의 자료실로 접근완료")

    #iframe에 접근
    driver.switch_to.frame("tool_content")
    driver.implicitly_wait(10)
    
    #최근 강의 클릭
    driver.find_element(By.XPATH,'//*[@id="xn-course-resource"]/div[1]/div/div/div[2]/div[3]/div/div').click()
    print("최근 강의로 접근중...")
    driver.implicitly_wait(10)
    print("최근 강의로 접근 완료")
    
    #강의자료 다운로드
    driver.find_element(By.XPATH,'//*[@id="xn-resource-view"]/div/div[3]/div[2]/p/a').click()
    print("다운로드 중...")
    driver.implicitly_wait(10)
    print("다운로드 완료")
```
